```r
# Recode the status variable into 1's and 0's'
## change y-axis to a numerical value rather than a categorical variable

mydata$cancer_status <- recode(mydata$status, 'M' = 1, 'B' = 0)

table(mydata$status)
table(mydata$cancer_status)
```

```r
mydata %>%
ggplot() +
  aes( x = radius, y = cancer_status) +
  geom_jitter(aes(col = status), height = 0.05, width = 0.01, alpha = 0.5)
```

```r
# prompt: can you add the logistic regression curve into my plot and col the curve red

mydata %>%
  ggplot() +
  aes(x = radius, y = cancer_status) +
  geom_jitter(aes(col = status), height = 0.05, width = 0.01, alpha = 0.5) +
  geom_smooth(method = "glm", method.args = list(family = "binomial"), se = FALSE, color = "red")
```

```r
mymodel <- glm(cancer_status ~ radius, data = mydata, family = binomial)
summary(mymodel)
```

Fitted curve:

$ \hat p = \frac{e^{-15.2+1.03radius}}{1 + e^{-15.2+1.03radius}}$

where $\hat p$ is the predicted probability from this model.

How to interpret the 'slope'?

- First, acknowledge the sign of the 'slope' parameter. In this case, the positive sign of the 'slope' indicates that probability increases as we move from small to large radius values.

- In order to think about the rate of change in probability we need to make some judgements on the x values we use.

```r
mynewdata <- data.frame(radius = c(14,15, 20,21))

mypredictions <- predict(mymodel, newdata = mynewdata, type = 'response')

mypredictions
## predictions are wrong because they are not probabilites (between 1 and 0), they are logs of our probabilites
## type = 'response' in our predictions fix this
```
